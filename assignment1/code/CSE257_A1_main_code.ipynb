Question 2

```python
# Implementation for Q2
def f(x1, x2):
    return x1*x1 - x1*x2 + 3*x2*x2 + 5

def df_dx1(x1, x2):
    return 2*x1 - x2

def df_dx2(x1, x2):
    return -x1 + 6*x2
```

Question 2.1 - Gradient Descent with constant step size $\alpha=0.3$

```python
# Gradient Descent with alpha=0.3
max_itr = 50
alpha = 0.3

f_val = np.zeros((max_itr+1,1))
x1_val = np.zeros((max_itr+1,1))
x2_val = np.zeros((max_itr+1,1))

[x1, x2] = [2, 2]
f_val[0,0] = f(x1, x2)
x1_val[0,0] = x1
x2_val[0,0] = x2

for itr in range(max_itr):
    p1 = -df_dx1(x1, x2)
    p2 = -df_dx2(x1, x2)

    x1 = x1 + alpha*p1
    x2 = x2 + alpha*p2

    x1_val[itr+1,0] = x1
    x2_val[itr+1,0] = x2
    f_val[itr+1,0] = f(x1, x2)

print("Minimum f(x1,x2) = ", round(f_val[max_itr,0],5), " achieved at (x1,x2) = ", (round(x1_val[max_itr,0],5), round(x2_val[max_itr,0],5)))
# print(f_val[20,0], f_val[50,0])
fig = plt.figure(figsize=(8,5))
plt.plot(range(max_itr+1), f_val)
plt.title('Function Minimzation')
plt.xlabel('iteration')
plt.ylabel('f(x1, x2)')
plt.grid(linestyle = '--')
plt.show()
plt.savefig("GD_2_1_loss.png")

fig = plt.figure(figsize=(8,5))
plt.plot(x1_val, x2_val)
for i in range(max_itr):
    x1_len = x1_val[i+1,0] - x1_val[i,0]
    x2_len = x2_val[i+1,0] - x2_val[i,0]
    plt.arrow(x1_val[i,0], x2_val[i,0], x1_len, x2_len, width=0, head_width = 0.05, ec='red')
plt.title('Function Minimization')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(linestyle = '--')
plt.show()
plt.savefig("GD_2_1_x1x2.png")
```

Question 2.2 - Gradient Descent with Optimal Step size in each step

```python
# Gradient Descent with Optimal Step Size alpha
max_itr = 20

f_val = np.zeros((max_itr+1, 1))
x1_val = np.zeros((max_itr+1, 1))
x2_val = np.zeros((max_itr+1, 1))

[x1, x2] = [2, 2]
f_val[0,0] = f(x1, x2)
x1_val[0,0] = x1
x2_val[0,0] = x2

for itr in range(max_itr):
    p1 = -df_dx1(x1, x2)
    p2 = -df_dx2(x1, x2)
    alpha_opt = -0.5*(2*x1*p1 - x1*p2 - x2*p1 + 6*x2*p2)/(p1*p1 - p1*p2 + 3*p2*p2)
    assert(alpha_opt >= 0)

    x1 = x1 + alpha_opt*p1
    x2 = x2 + alpha_opt*p2

    x1_val[itr+1,0] = x1
    x2_val[itr+1,0] = x2
    f_val[itr+1,0] = f(x1, x2)

print("Minimum f(x1,x2) = ", round(f_val[max_itr,0],5), " achieved at (x1,x2) = ", (round(x1_val[max_itr,0],5), round(x2_val[max_itr,0],5)))
fig = plt.figure(figsize=(8,5))
plt.plot(range(max_itr+1), f_val)
plt.title('Function Minimzation')
plt.xlabel('iteration')
plt.ylabel('f(x1, x2)')
plt.grid(linestyle = '--')
plt.show()

fig = plt.figure(figsize=(8,5))
plt.plot(x1_val, x2_val)
for i in range(max_itr):
    x1_len = x1_val[i+1,0] - x1_val[i,0]
    x2_len = x2_val[i+1,0] - x2_val[i,0]
    plt.arrow(x1_val[i,0], x2_val[i,0], x1_len, x2_len, width=0, head_width = 0.05, ec='red')
plt.title('Function Minimzation')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(linestyle = '--')
plt.show()
```

```python
# Question 4 - Gradient Descent to minimize L(A,B)
max_itr = 50
alpha = 1e-2
L_val = np.zeros((max_itr+1, 1))

X_data = np.transpose(np.array([[1,2,3],[1,2,3],[-1,2,-1]]))
Y_data = np.transpose(np.array([[-1,-3,1],[1,-3,1],[0,1,2]]))

print("X_data: \n", X_data, "\n Y_data: \n", Y_data)
```

Question 4.2 - Gradient Descent to find solution $A,B$ to minimize $L(A,B)$

```python
def compute_dL_dA(A, B):
    term1 = np.zeros((3,3))
    term2 = np.zeros((3,3))
    term3 = np.zeros((3,3))

    for i in range(3):
        term1 += np.matmul(np.matmul(A, np.reshape(X_data[:,i], (3,1))), np.reshape(X_data[:,i], (1,3)))
        term2 += np.matmul(B, np.reshape(X_data[:,i], (1,3)))
        term3 += np.matmul(np.reshape(Y_data[:,i], (3,1)), np.reshape(X_data[:,i], (1,3)))

    return 2*(term1 + term2 - term3)
```

```python
def compute_dL_dB(A, B):
    term1 = np.zeros((1,3))
    term2 = np.zeros((1,3))

    for i in range(3):
        term1 += np.reshape(np.matmul(np.transpose(np.reshape(X_data[:,i], (3,1))), np.transpose(A)), (1,3))
        term2 += np.reshape(np.transpose(Y_data[:,i]), (1,3))

    return np.transpose(2*(term1 - term2 + 3*np.transpose(B)))
```

```python
def compute_loss(A,B):
    fx = np.zeros((3,3))
    loss = 0
    for i in range(3):
        fx[:,i] = np.reshape(np.matmul(A, np.reshape(X_data[:,i],(3,1))) + B, (3))
        loss += np.square(np.linalg.norm(fx[:,i] - Y_data[:,i]))
    return loss
```

```python
# initialize the parameters with random values
# A = np.random.rand(3,3)
# B = np.random.rand(3,1)
A = np.random.normal(0, 0.1, (3,3))
B = np.random.normal(0, 0.1, (3,1))

print("Initial parameters: \n A = \n", A, "\n B = \n", B, "\n")

L_val[0,0] = compute_loss(A,B)

for itr in range(max_itr):
    dL_dA = compute_dL_dA(A, B)
    dL_dB = compute_dL_dB(A, B)

    pA = -dL_dA
    pB = -dL_dB

    A = A + alpha*pA
    B = B + alpha*pB

    L_val[itr+1,0] = compute_loss(A,B)

print("Minimum L(A,B) =", round(L_val[max_itr,0],5), "after", max_itr, "iterations of gradient descent with alpha =", alpha, "\n A = \n", A, "\n B = \n", B)
fig = plt.figure(figsize=(8,5))
plt.plot(range(max_itr+1), L_val)
plt.title('Function Minimzation')
plt.xlabel('iteration')
plt.ylabel('L(A,B)')
plt.grid(linestyle = '--')
plt.show()
```
